```python
import imdb
import pandas as pd
```

```python
ia = imdb.IMDb()
```

```python
buffy_search = ia.search_movie('buffy the vampire slayer')
buffy_search
```

```python
angel_search = ia.search_movie('angel')
angel_search
```

```python
buffy_ep = ia.get_movie_episodes('0118276')
angel_ep = ia.get_movie_episodes('0162065')
```

```python
#can I create a list of these items? 
master_list = [buffy_ep,angel_ep] # yes you can 
```

```python
#created lots of loops to scale the process
master_df = pd.DataFrame()

for x in range(len(master_list)):
    for y in master_list[x]['data']['episodes'].keys():
        for z in master_list[x]['data']['episodes'][y].keys():
            temp_dict = master_list[x]['data']['episodes'][y][z].data
            temp_df = pd.DataFrame(temp_dict, index = [0])
            temp_df['episode of'] = temp_df['episode of'].astype(str)
            master_df = master_df.append(temp_df)
            #print(f"Order {x} Season{y} episode{z}")
```

```python
master_df.info()
```

```python
#reset index for pandas
master_df = master_df.reset_index(drop=True)
```

```python
#confirm any unique values in kind column
#print(master_df['kind'].unique())
#drop kind column
master_df = master_df.drop('kind',axis = 1)
```

```python
#I don't consider either unaired pilot cannon, so I'm dropping them
master_df[master_df['episode']==0]
```

```python
#actual drop code
master_df = master_df.drop(master_df.index[[0,167]])
```

```python
#clean up (TV Series from Episode of)

#remove everything to the right of the ( by subsetting the first part of the string
master_df['episode of'] = master_df['episode of'].str.split("(").str[0]

#removes the trailing white space. Can't see it, but I know it is there
master_df['episode of'] = master_df['episode of'].str.strip()
```

```python
#sort by the orginal air date

#month has a period at the end of it, so need to remove it
master_df['original air date'] = master_df['original air date'].str.replace('.','')
```

```python
master_df.tail()
```

```python
#convert to date time. 
#using pd.to_datetime, pandas will automatiically figure it out
master_df['original air date'] = pd.to_datetime(master_df['original air date'])
```

```python
#sort from oldest to newest
master_df = master_df.sort_values(by = ['original air date','season','episode'])
```

```python
map_dict = {
    'Buffy the Vampire Slayer':'buffy',
    'Angel':'angel'
}
```

```python
#use mapping function for quick relabel
master_df['show_code']=master_df['episode of'].map(map_dict)
```

```python
master_df = master_df.reset_index(drop=True)
```
