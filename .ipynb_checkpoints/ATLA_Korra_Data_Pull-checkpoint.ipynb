```python
#pull title list and movie_id list to create pandas df

id_list = []

for i in range(len(tv_show)):
    temp_list = tv_show[i].getID()
    id_list.append(temp_list)

title_list = []

for i in range(len(tv_show)):
    temp_list = str(tv_show[i]) #needs to be a str otherwise pull it in as a IMDB.movie.movie object...
    title_list.append(temp_list)
```

```python
#allows me to get movies and their ids into a pandas df for easier searching and manipulation in the future
new_df = pd.DataFrame()

new_df['id'] = id_list
new_df['title'] = title_list
```

```python
lok_ep = ia.get_movie_episodes('1695360')
atla_ep = ia.get_movie_episodes('0417299')
tdf_ep = ia.get_movie_episodes('8688814') # confirmed this is the correct series
```

```python
#can I create a list of these items? 
master_list = [lok_ep,atla_ep,tdf_ep] # yes you can 
```

```python
## because I subset to data and then episodes, I don't need this step
"""
master_ep_list = []

for x in range(len(master_list)):
    temp_list = master_list[x]['data'].get('episodes')
    master_ep_list.append(temp_list)
"""

#master_ep_list
```

```python
#created lots of loops to scale the process
master_df = pd.DataFrame()

for x in range(len(master_list)):
    for y in master_list[x]['data']['episodes'].keys():
        for z in master_list[x]['data']['episodes'][y].keys():
            temp_dict = master_list[x]['data']['episodes'][y][z].data
            temp_df = pd.DataFrame(temp_dict, index = [0])
            temp_df['episode of'] = temp_df['episode of'].astype(str)
            master_df = master_df.append(temp_df)
            #print(f"Order {x} Season{y} episode{z}")
```

```python
#reset index for pandas
master_df = master_df.reset_index(drop=True)
```

```python
#confirm any unique values in kind column
print(master_df['kind'].unique())
#drop kind column
master_df = master_df.drop('kind',axis = 1)
```

```python
master_df = master_df.drop(master_df.index[114])
```

```python
test = list(master_df['episode of'].unique())
```

```python
#clean up (TV Series from Episode of)

#remove everything to the right of the ( by subsetting the first part of the string
master_df['episode of'] = master_df['episode of'].str.split("(").str[0]

#removes the trailing white space. I know it is there
master_df['episode of'] = master_df['episode of'].str.strip()
```

```python
#sort by the orginal air date

#month has a period at the end of it, so need to remove it
master_df['original air date'] = master_df['original air date'].str.replace('.','')
```

```python
master_df.head()
```

```python
#convert to date time. 
#using pd.to_datetime, pandas will automatiically figure it out
master_df['original air date'] = pd.to_datetime(master_df['original air date'])
```

```python
#sort from oldest to newest
master_df = master_df.sort_values(by = ['original air date','season','episode'])
```

```python
master_df = master_df.reset_index(drop=True)
```
